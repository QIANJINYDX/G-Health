```python
import os
import sys
import glob
import shutil
import pandas as pd
import random
from PIL import Image
```

移动Breast_cancer

```python
source_dict = {
    'breast_cancer': glob.glob("/home/dxye/Program/PhysicalExaminationAgent/client/risk_assessment/data/Img/Breast_cancer/**/*.png", recursive=True),
    'chest_cancer': glob.glob("/home/dxye/Program/PhysicalExaminationAgent/client/risk_assessment/data/Img/chest_cancer/**/*.png", recursive=True),
    'eye_disease': glob.glob("/home/dxye/Program/PhysicalExaminationAgent/client/risk_assessment/data/Img/eye_disease/**/*.png", recursive=True),
    'skin_cancer': glob.glob("/home/dxye/Program/PhysicalExaminationAgent/client/risk_assessment/data/Img/Skin_Cancer/**/*.jpg", recursive=True)
}
```

```python
target_root = "/home/dxye/Program/PhysicalExaminationAgent/client/risk_assessment/model_router/data"
os.makedirs(target_root, exist_ok=True)
```

```python
records = []
# 遍历每个类别
for label, paths in source_dict.items():
    sample_paths = random.sample(paths, min(100, len(paths)))  # 随机选最多100张
    target_dir = os.path.join(target_root, label)
    os.makedirs(target_dir, exist_ok=True)

    for i, src_path in enumerate(sample_paths):
        try:
            with Image.open(src_path) as img:
                # 统一命名
                filename = f"{label}_{i:03d}.png"
                target_path = os.path.join(target_dir, filename)
                img.convert("RGB").save(target_path, format="PNG")

                # 保存CSV记录（使用相对路径）
                rel_path = os.path.relpath(target_path, target_root)
                records.append((rel_path, label))
        except Exception as e:
            print(f"跳过 {src_path}: {e}")
```

```python
df = pd.DataFrame(records, columns=["image_path", "label"])
csv_path = os.path.join(target_root, "metadata.csv")
df.to_csv(csv_path, index=False)
```
